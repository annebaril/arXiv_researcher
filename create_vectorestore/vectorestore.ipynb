```python
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
import json

# Étape 1 : Récupérer les articles d'arXiv via l'API
path_json = "../collect_arxiv_data/articles_arxiv.json"
with open(path_json, "r", encoding="utf-8") as file:
    articles = json.load(file)
```

```python
articles
```

```python
# Extraire les titres et résumés des articles
documents = []
for entry in articles:
    title = entry['titre']
    summary = entry['résumé']
    documents.append(f"{title}: {summary}")

# Étape 2 : Utiliser le modèle SentenceTransformer pour obtenir les embeddings
embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-mpnet-base-v2')

# Étape 3 : Créer un vector store Chroma avec les résumés des articles
vectorstore = Chroma.from_texts(
    texts=documents,
    embedding=embeddings,
    persist_directory="chroma_db"  # Le dossier où les données seront persistées
)

# Sauvegarder le vector store sur disque
vectorstore.persist()

print("Chroma vector store a été créé et sauvegardé dans 'chroma_db' avec les articles d'arXiv.")
```
